```python
import math

K = [
    (19, 16.0, 0.018, 0.25),
    (15, 8.4, 0.017, 0.3),
    (19, 17.0, 0.025, 0.25),
    (21, 14.9, 0.056, 0.3),
]


def f(x):
    return a - b * math.exp(-c * x)


for k in K:
    a, b, c, d = k
    print(f"Параметры культуры: a={a} b={b} c={c} d={d}")
    for i in range(0, 239, 10):
        print(f"{i} {f(i)}")

```
