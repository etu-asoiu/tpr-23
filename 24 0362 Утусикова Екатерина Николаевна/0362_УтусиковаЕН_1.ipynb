В цехах N1 и N2 предприятия производится продукт Y, который в дальнейшем используется в качестве исходного материала для производства изделий в цехе N3. Суммарная производительность цехов N1 и N2 зависит от вложения дополнительных средств X. При работе цехов N1 и N2 в течение одного месяца эта зависимость может быть приближенно представлена в виде функций:

• N1: y = 5 + (x + 40)^1/3;

• N2: y = 7 + (x + 30)^1/3;

Функции остатка средств в течение месяца:

• N1: 0.83x;

• N2: 0.77x.

Средства,выделяемые на оба цеха в течение квартала (3 месяца), составляют 174 единиц; перераспределение производится помесячно.

Требуется распределить средства на планируемый квартал с целью получения максимального количества продукта Y.

**Формализация**

Дано: X = 174

Для решения данной задачи необходимо построить функцию суммарной производительности цехов N1 и N2 за весь квартал и определить ее максимум.

За один месяц цех N1 производит:

y1 = 5 + (x11 + 40)^1/3

За три месяца цех N1 производит:

y = 15 + (x11 + 40)^1/3 + (x12 + 40)^1/3 + (x13 + 40)^1/3

Аналогично для цеха N2:

y2 = 7 + (x21 + 30)^1/3

y = 21 + (x21 + 30)^1/3 + (x22 + 30)^1/3 + (x23 + 30)^1/3

Суммарная производительность цехов за квартал будет:

Y = y1 + y2 = 36 + (x11 + 40)^1/3 + (x12 + 40)^1/3 + (x13 + 40)^1/3 + (x21 + 30)^1/3 + (x22 + 30)^1/3 + (x23 + 30)^1/3

X = x11 + x21

Остаток средств после месяца использования для цеха N1:

K1 = 0.83x

Остаток средств после месяца использования для цеха N2:

K2 = 0.77x

Общий остаток средств после первого месяца:

k1 = K1 + K2 = 0.83 * x11 + 0.77 * x21 = x12 + x22

На второй месяц остается средств:

k2 = 0.83 * x12 + 0.77 * x22 = x13 + x23

На третий месяц остается средств:

k3 = 0.83 * x13 + 0.77 * x23

Основное функциональное уравнение динамического программирования для решения задачи:

Wi (ki) =   max ⁡{ 12 + (x + 5)^(1/3) + (x + 7)^(1/3) + W(i+1) (0.83x + 0.77(k-x))}, где xi <= ki

**Решение**

Находим оптимальное управление на первом шаге x11 = 96.67, показывающее, сколько средств надо вкладывать в первый цех, и соответствующую максимальную производительность за квартал W1 = 64.386, а также количество средств вкладываемые во второй цех:

Таким образом, можно сформулировать следующие рекомендации по оптимальному распределению средств. Из имеющегося в начале квартала запаса средств k = 174 усл. ед. и остающихся средств в конце каждого месяца нужно вкладывать по месяцам в цеха I и II следующие суммы:





изображение.png

При таком планировании будет получена максимальная производительность за 1 квартал года  W1max = 64.386 ≈ 64 ед.

Остаток средств на конец квартала: 51.22 * 0.83 + 61.23 * 0.77 = 89.66 ≈ 90 ед.


```python
import numpy as np
#Производительность цеха N1

def c(x):
  v = 5 + (x + 40)**(1/3)
  return v
#Производительность цеха N2

def b(y):
  v = 7 + (y + 30)**(1/3)
  return v
#Общая производительность

def w(k, x):
  y = c(x) + b(k - x)
  return y
#Функция подсчета остатков после каждого периода

def phi(k, x):
  y = 0.83* x + 0.77*(k - x)
  return y

def W(k, ks, Ws):
  xs = np.arange(0, k, 0.01)
  next_k = phi(k, xs)
  vals = w(k, xs) + Ws[np.searchsorted(ks[:9], next_k, 'left')]
  besti = np.argmax(vals)
  v = vals[besti]
  x = xs[besti]
  return v, x

"""
def Wx(k, xs, ks, Ws):
  next_k = phi(k, xs)
  vs = w(k, xs) + Ws[np.searchsorted(ks, next_k, 'left')]
  return vs
"""

k_4 = np.linspace(0.83**3, 0.77**3, 10)*174
W_4 = np.zeros_like(k_4)
print ("Остатки:", k_4)
print("----")
print ("w4", W_4)
print ("k4", k_4)


x1_values = []
y1_values = []

k_3 = np.linspace(0.83**2, 0.77**2, 10)*174
W_3 = np.zeros_like(k_3)
x_3 = np.zeros_like(k_3)
print("")
print("Третий месяц")
for i in range(len(k_3)):
  W_3[i], x_3[i] = W(k_3[i], k_4, W_4)
  print("Производительность:          ", "Вкладываемые средства:           ", "Остатки:")
  print ( round(W_3[i], 2),"                       ", round(x_3[i], 2),"                            ", round(k_3[i], 2))

k_2 = np.linspace(0.83, 0.77, 10)*174
W_2 = np.zeros_like(k_2)
x_2 = x_3
print("")
print("Второй месяц")
for i in range(len(k_2)):
  W_2[i], x_2[i] = W(k_2[i], k_3, W_3)
  print("Производительность:          ", "Вкладываемые средства:           ", "Остатки:")
  print ( round(W_2[i], 2),"                       ", round(x_2[i], 2),"                            ", round(k_2[i], 2))


k_1 = np.linspace(0.83**0, 0.77**0, 10)*174
W_1 = np.zeros_like(k_1)
x_1 = x_2
print("")
print("Первый месяц")
for i in range(len(k_1)):
  W_1[i], x_1[i] = W(k_1[i], k_2, W_2)
  print("Производительность:          ", "Вкладываемые средства:           ", "Остатки:")
  print ( round(W_1[i], 2),"                       ", round(x_1[i], 2),"                            ", round(k_1[i], 2))
```
